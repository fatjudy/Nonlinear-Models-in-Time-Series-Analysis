```python
import numpy as np
import pandas as pd
import matplotlib as plt
```

# Import Data and Select Variables

```python
Data_raw = pd.read_excel('data/PredictorData2020.xlsx', sheet_name = 'Quarterly')
Data_raw.head()
```

```python
Data_1 = Data_raw[Data_raw['yyyyq'] > 19263]
Data_1.head()
```

```python
Data_2 = Data_1.drop(['csp', 'CRSP_SPvwx'], axis=1)
Data_2.head()
```

# Construct Dependent Variable: EQP

```python
df_temp = pd.DataFrame(Data_2)
df_temp.index = np.arange(len(df_temp))
df_temp.head()
```

```python
df = df_temp.iloc[1:,] # df start from Jan of 1927 
df.head()
```

```python
# Market return: simple return on the S&P 500 (including dividends)
rt = df['CRSP_SPvw']
rt.head()
```

```python
# Simple return on a risk-free bill
df_lag = df_temp[:-1]
df_lag.index = np.arange(1, len(df_lag)+1)
df_lag.head()
```

```python
it_lag = df_lag['Rfree']
it_lag.head()
```

```python
# log of Equity Premium:ln(1+rt) - ln(1+i(t-1))
EQP = pd.Series(np.log(1+rt) - np.log(1+it_lag), name = 'EQP')
EQP.head()
```

# Construct Independent Variables

```python
# Dividend Price Ratio
dp = pd.Series(np.log(df['D12']) - np.log(df['Index']), name = 'DP')
dp.head()
```

```python
# Dividend Yield
dy = pd.Series(np.log(df['D12']) - np.log(df_lag['Index']), name = 'DY')
dy.head()
```

```python
# Earnings Price Ratio
ep = pd.Series(np.log(df['E12']) - np.log(df['Index']), name = 'EP')
ep.head()
```

```python
# Divident-payout Ratio
de = pd.Series(np.log(df['D12']) - np.log(df['E12']), name = 'DE')
de.head()
```

```python
# dfy: Default Yield Spread
dfy = pd.Series(df['BAA'] - df['AAA'], name = 'DFY')
dfy.head()
```

```python
# dfr: Default Return Spread
dfr = pd.Series(df['corpr'] - df['ltr'], name = 'DFR')
dfr.head()
```

```python
tms = pd.Series(df['lty'] - df['tbl'], name = 'TMS')
tms.head() # This is different from Deshui's dataset
```

# Data Frame for analysis

```python
Preds = pd.concat([df['yyyymm'], EQP, dp, dy, ep, de, df['svar'], df['b/m'], df['ntis'], df['tbl'], df['lty'], 
                df['ltr'], tms, dfr, dfy, df['infl']], axis=1)
Preds = pd.DataFrame(Preds)
Preds.head()
```

```python
Preds.to_excel("Ying_Data_2019.xlsx")  
```
