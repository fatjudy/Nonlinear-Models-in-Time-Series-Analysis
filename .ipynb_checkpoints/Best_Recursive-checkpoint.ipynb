```python
combine_list = name_list +fun_list
for i in combine_list:
    fig = plt.figure(figsize=(8,8))
    dfp = pd.read_excel('combination_plot.xlsx', sheet_name = i, index_col = 0)

    dfp['nls_SM'].plot(legend = True)
    plt.hlines(y = 0, xmin = 0, xmax = 201812, colors='r')
    plotname = i+'_'+'combination'
    plt.title(plotname)
    plt.tight_layout()
    filename = plotname + '_1952_q_' + 'SM' + '.png'
#         plt.show()
    plt.savefig(fname = filename)
```
